# LLR Scaling for Stable Vs. Unstable parts
LLR of BSC scales as a function of $p$ as:
$ \log((1-p)/p) $

If $p$ is multiplied by a factor $c$ (to express stable vs. unstable part), how does the llr change with $c$?

Denote $f=cp$, then the LLR scales as:

$$ \log\left(\frac{1-f}{f}\right) = \log\left(\frac{1-cp}{cp}\right) $$

```python
import numpy as np
p = 0.05
c = np.arange(0.1,6,0.01)
f = c*p
llr = np.log((1-f)/f)
```

```python
plt.plot(c,llr)
plt.show()
```

```python
delta_llr = llr - np.log((1-p)/p)
plt.plot(c,delta_llr)
plt.show()
```

```python
llr_ratio = llr/np.log((1-p)/p)
plt.plot(c,llr_ratio)
plt.show()
```
